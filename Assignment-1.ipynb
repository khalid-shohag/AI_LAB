```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

# Generation of input data

```python
X = np.arange(-10, 10, 0.1)
X[:5], len(X)
```

# y = wx + b; where 'w' and 'b' are constants

```python
W = random.randint(1, 20)
b = random.randint(1, 20)
W, b

#Output
y = W*X + b
plt.plot(X, y)
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression')
```

# y = x^2

```python
y = np.power(X, 2)
plt.plot(X, y)
plt.xlabel('X')
plt.ylabel('y')
plt.title('Quadratic Regression')
```

# y = 1/ (1 + e^(-x))

```python
y = 1/(1 + np.exp(-X))
plt.plot(X, y)
plt.xlabel('X')
plt.ylabel('y')
plt.title('Sigmoid')
```

# y = (e^x - e^(-x)) / (e^x + e^(-x))


```python
y = (np.exp(X) - np.exp(-X)) / (np.exp(X) + np.exp(-X))
plt.plot(X, y)
plt.xlabel('X')
plt.ylabel('y')
plt.title('Hyperbolic Tangent')
```

# y = g(f(x)) where u = f(x) = wx + b and y = g(u) = 1/ (1 + e^(-u))

```python
f = W*X + b
y = 1/(1 + np.exp(-f))
plt.plot(X, y)
plt.xlabel('X')
plt.ylabel('y')
plt.title('Logistic Regression')
```

# y = g(f(x)) where f(x) = wx + b, u = f(x), and g(u) = (e^u - e^(-u)) / (e^u + e^(-u))

```python
f = W*X + b
y = (np.exp(f) - np.exp(-f)) / (np.exp(f) + np.exp(-f))
plt.plot(X, y)
plt.xlabel('X')
plt.ylabel('y')
plt.title('Hyperbolic Tangent')
```
